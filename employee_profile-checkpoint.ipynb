# Company Profile Analysis
This project covers the entire process—from manipulating and grouping the data to in-depth analysis and visualization using company profile dataset.
* Finding the average salary by year.
* Showing the ratio of remote or non-remote employees.
* Visualize the company size based on the years. 
* Printing out the top 5 job titles.
* Finding the distribution of salary by company size. 
* Creating a histogram showing the distribution of salaries by company size.
* Finding the experience level of employees.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
emp = pd.read_csv("Employee_profile.csv")
emp.head()
```

```python
emp.shape
```

```python
emp.drop('Unnamed: 0', axis = 1, inplace = True)
emp
```

```python
df1 = emp.groupby('work_year')['salary_in_usd'].mean().round(2)
df1
```

```python
df1.index
```

```python
df1.values
```

```python
data = {'work_year' : df1.index, 'average_salary' : df1.values}
df1 = pd.DataFrame(data)
df1
```

```python
df1['average_salary'] = (df1['average_salary']/1000).round(2)
df1
```

```python
ax = df1.plot(kind = 'bar', x = 'work_year', y = 'average_salary', color='b', legend = True)
ax.bar_label(ax.containers[0], labels = df1['average_salary'].map('{:.2f}k'.format), padding = 3)

plt.subplots_adjust(top=1.2)
ax.legend(['Average Salary'], loc = 'upper left')

plt.xlabel('Work Year')
plt.ylabel('Average Salary')
plt.title('Average Salaries by Years', c='purple')
plt.show()
```

```python
df2 = emp.remote_ratio.value_counts()
df2
```

```python
labels = ['Fully Remote', 'No Remote', 'Partially Remote'] 
values = df2.to_list()
values
```

```python
z = plt.bar(labels, values, width = 0.5, color = 'lavender')
plt.bar_label(z, labels = values)
plt.title('Employees in a Remote Position', c='purple')
plt.ylabel('Counts')
plt.show()
```

```python
df3 = emp.company_size.value_counts()
df3
```

```python
df3.index 
```

```python
df4 = emp.job_title.value_counts().head()
df4
```

```python
plt.figure(figsize = (10,5))
# plt.bar(x = df4.index, height=df4.values)
sns.barplot(x = df4.index, y = df4.values, hue = df4.index, palette = 'viridis')
plt.title('Top 5 Job Titles', c='purple')
plt.ylabel('Counts')
plt.xticks(rotation=30)
plt.show()
```

```python
df5 = emp[["salary_in_usd", "company_size"]]
df5
```

```python
df5[df5['company_size'] == 'S']
```

```python
S = df5[df5['company_size'] == 'S']
M = df5[df5['company_size'] == 'M']
L = df5[df5['company_size'] == 'L']
company_label = ['Medium', 'Large', 'Small']

sal_mean = [S['salary_in_usd'].mean(), M['salary_in_usd'].mean(), L['salary_in_usd'].mean()]

sal_mean
```

```python
label_change = np.round([x/1000 for x in sal_mean], 2)
label_change = list(map(str, label_change))
label_change = [x + 'K' for x in label_change]
label_change
```

```python
z = plt.bar(company_label, sal_mean, color = 'purple', alpha = 0.5)
plt.bar_label(z, label_change, alpha = 0.8)
plt.title('Distribution of Salary by Company Size', fontstyle='italic', fontsize=15, c ='purple')
plt.xlabel('Size of Company')
plt.ylabel('Salary in USD')
plt.show()
```

```python
sns.histplot(M['salary_in_usd'], label = 'Medium', kde=True)
sns.histplot(L['salary_in_usd'], label = 'Large', kde=True)
sns.histplot(S['salary_in_usd'], label = 'Small', kde=True)
plt.legend()
plt.title('Distribution of Salary by Company Size', fontstyle='italic', fontsize=15, c ='purple')
plt.xlabel('Salary in USD')
plt.show()
```

```python
exp_map = { 'SE' : 'Senior', 'MI' : 'Middle', 'EN' : 'Entry level', 'EX' : 'Executive'}
```

```python
emp['experience_level'] = emp['experience_level'].replace(exp_map)

df6 = emp.experience_level.value_counts()
df6
```

```python
labels = df6.index.to_list()
labels
```

```python
values = df6.values
values
```

```python
plt.figure(figsize = (5,10))
plt.pie(x = values, labels = labels, autopct = '%0.2f%%', shadow = True)
plt.title('Experience level')
plt.show()
```
